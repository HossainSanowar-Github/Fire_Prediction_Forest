# 2. Connect MongoDb Server

```python
# Connect the database to a MongoDB Atlas Cluster using Secure Authentication with a User Name and Password.
client =MongoClient("mongodb+srv://mongodb:mongodb@cluster0.qvijt.mongodb.net/?retryWrites=true&w=majority")
db = client.test
print(db)

# Create Database and specify database's name.
db = client.get_database('Algerian_fires_db')
#Establish a collection
records = db.fire_records
```

# 3. Load CSV file from localhost

```python
# Create a Dataframe using Pandas and read the dataset
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
dataset.head()
```

```python
dataset.count()
```

# 4. Convert DataFrame to Dictionary and Upload to MongoDB

```python
#Convert Dataframe to Dictionary since MongoDB stores information as documents/records.
data = dataset.to_dict(orient = 'records')
```

```python
# Insert data from the dataset into the "fire records" collection in MongoDB.
db.fire_records.insert_many(data)
print("All the Data has been Exported to MongoDB Successfully")
```

# 5. Load Data from Mongo DB

```python
#Load all records from MongoDB using find()
all_records = records.find()
print(all_records)
```

```python
#Convert Cursor Object into list
list_cursor = list(all_records)
```

```python
#Convert list into Dataframe
df = pd.DataFrame(list_cursor)
df.drop('_id', axis=1, inplace=True)
df
```
